```python
MySQL_required_columns = MySQL[["Operations_in_Query", "95_lower_bound_of_confidence_interval_for_t_distribution", "95_upper_bound_of_confidence_interval_for_t_distribution"]]
```

```python
MongoDB_required_columns = MongoDB[["Operations_in_Query", "95_lower_bound_of_confidence_interval_for_t_distribution", "95_upper_bound_of_confidence_interval_for_t_distribution"]]
```

```python
MySQL_required_columns["MySQL_y_label"] = MySQL_required_columns[["95_lower_bound_of_confidence_interval_for_t_distribution", "95_upper_bound_of_confidence_interval_for_t_distribution"]].mean(axis=1)
```

```python
MongoDB_required_columns["MongoDB_y_label"] = MongoDB_required_columns[["95_lower_bound_of_confidence_interval_for_t_distribution", "95_upper_bound_of_confidence_interval_for_t_distribution"]].mean(axis=1)
```

```python
MySQL_required_columns['Operations_in_Query'] = MySQL_required_columns['Operations_in_Query'].apply(lambda x: x.replace(" ", "").replace(".", "").replace(",,", ",").upper())
```

```python
MySQL_required_columns[["Operations_in_Query", "MySQL_y_label"]].head()
```

```python
MongoDB_required_columns['Operations_in_Query'] = MongoDB_required_columns['Operations_in_Query'].apply(lambda x: x.replace(" ", "").replace(".", "").replace(",,", ",").upper())
```

```python
MongoDB_required_columns[["Operations_in_Query", "MongoDB_y_label"]].head()
```

```python
MySQL_required_columns["All_query_operations"] = MySQL_required_columns["Operations_in_Query"] + "," + MongoDB_required_columns["Operations_in_Query"]
```

```python
MySQL_required_columns["Suggested_DB"] = np.where(MySQL_required_columns["MySQL_y_label"] < MongoDB_required_columns["MongoDB_y_label"], "MySQL", "MongoDB")
```

```python
final_df = MySQL_required_columns[["All_query_operations", "Suggested_DB"]]
```

```python
final_df
```

```python
final_df = pd.concat([final_df.drop('All_query_operations', 1), final_df['All_query_operations'].str.get_dummies(sep=",")], 1)
```

```python
final_df.head()
```

```python
final_df.loc[final_df['Suggested_DB'] == "MySQL", 'Suggested_DB'] = 0 
final_df.loc[final_df['Suggested_DB'] == "MongoDB", 'Suggested_DB'] = 1
```

Now, let's build our model and get the accuracy

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
import itertools
from sklearn.metrics import accuracy_score
```

```python
Y = final_df["Suggested_DB"]
X = final_df.drop(["Suggested_DB"], axis = 1)
```

```python
seed = 2021
test_size = 0.25
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
```

```python
# fit model no training data
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(list(y_test), predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
confusion = [[0, 0], [0, 0]]
confusion = np.add(confusion, confusion_matrix(list(y_test), predictions))
```

```python
### From http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html#sphx-glr-auto-examples-model-selection-plot-confusion-matrix-py
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        #print("Normalized confusion matrix")
    #else:
    #    print('Confusion matrix, without normalization')

    #print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
np.set_printoptions(precision=2)

# Plot confusion matrix for all 5 combined test sets
plt.figure(figsize=(8,8))
plot_confusion_matrix(confusion, classes=['MySQL', 'MongoDB'],
                      title='Confusion matrix')
plt.show()
```
